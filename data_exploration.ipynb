```python
import numpy as np

from utils import load_data, split_dataset, IMG_DIM
from visualisation import get_discrete_bins, plot_histogram, show_samples
```

```python
x, y = load_data()
```

```python
plot_histogram(y, 'Class distribution', 'class', 'number of cases', get_discrete_bins(y), True)
```

```python
number_of_white_pixels = np.sum(x.reshape(x.shape[0], IMG_DIM * IMG_DIM), axis=1)
plot_histogram(number_of_white_pixels, 'Number of white pixels distribution', 'number of white pixels on image', 'number of cases')

print('Images with biggest amount of white pixels')
show_samples(x[np.argsort(number_of_white_pixels)[-16:]])

print('Images with smallest amount of white pixels')
show_samples(x[np.argsort(number_of_white_pixels)[:16]])
```

```python
def get_samples(x, y, label, n_samples):
    possible_args = np.where(y == label)[0]
    selected_args = np.random.choice(possible_args, min(n_samples,  possible_args.shape[0]), False)
    return x[selected_args]

print('Examples of class instances')
for label in np.arange(y.max()):
    print(label)
    samples = get_samples(x, y, label, 16)
    show_samples(samples)
```

```python
x_train, x_test, y_train, y_test = split_dataset(x, y)
plot_histogram(y_train, 'Class distribution in used train set', 'class', 'number of cases', get_discrete_bins(y), True)
plot_histogram(y_test, 'Class distribution in used test', 'class', 'number of cases', get_discrete_bins(y), True)
```
